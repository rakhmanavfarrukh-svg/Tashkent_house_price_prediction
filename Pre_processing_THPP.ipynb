```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_data_08-02-2021.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[(df['size'] < 500) & (df['price'] < 1000000) ]
df.describe()
```

```python
#for size column
df['size'] = pd.to_numeric(df['size'], errors='coerce')
df.dropna(subset=['size'], inplace=True)
df['size'] = df['size'].round()
df['size'] = df['size'].astype('int64')
#for price column
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.dropna(subset=['price'], inplace=True)
df['price'] = df['price'].astype('int64')
#see changes
df.info()
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(df,test_size=0.2,random_state=50)
```

```python
housing=train_set.drop(['price','location'],axis=1)
housing_labels = train_set['price'].copy()
```

```python
housing.info()
```

```python
housing.head()
```

```python
housing_onehot = pd.get_dummies(housing['district']).astype(int)
housing_onehot
```

```python
housing = housing.drop('district', axis=1)
housing = pd.concat([housing, housing_onehot], axis=1)
```

```python
housing.head()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, size_ix = 0, 1
class CombinedAttributes(BaseEstimator, TransformerMixin):
    def __init__(self, add_size_per_room=True):
        self.add_size_per_room = add_size_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        rooms_per_size = X[:, size_ix] / X[:, rooms_ix]
        if self.add_size_per_room:
            return np.c_[X, rooms_per_size]
        else:
            return X
```

```python
attr_adder = CombinedAttributes()
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
housing_extra_attribs = attr_adder.transform(housing.values)
housing.insert(4, 'size_per_room', housing_extra_attribs[:, -1])
```
