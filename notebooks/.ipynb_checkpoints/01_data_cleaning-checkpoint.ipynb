# Tashkent House Price Prediction - Data Cleaning

## Objective
The goal of this notebook is to clean the raw dataset by handling missing values,
fixing data types, and preparing the dataset for further analysis.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Farruxbek\Downloads\housing_data_08-02-2021.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['size'].unique()
```

```python
df['price'].unique()
```

```python
# Convert to numeric (invalid - NaN)
df['size'] = pd.to_numeric(df['size'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
# Fill missing values
df['size'] = df['size'].fillna(df['size'].median())
df['price'] = df['price'].fillna(df['price'].median())
```

```python
# Remove unrealistic sizes  and prices
df = df[(df['size'] > 10) & (df['size'] < 500)]
df = df[(df['price'] > 10000) & (df['price'] < 500000)]
```

```python
df.to_csv('../data/cleaned_data.csv', index=False)
```

## Conclusion

- Missing values have been handled
- Data types have been corrected
- Dataset is now clean and ready for exploratory data analysis
