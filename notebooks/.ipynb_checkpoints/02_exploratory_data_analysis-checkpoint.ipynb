```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned dataset
df = pd.read_csv("../data/cleaned_data.csv")
print("Shape",df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Correlation Analysis

Only numeric features were used for correlation analysis,
as categorical variables cannot be directly used.


```python
df.corr(numeric_only=True)['price'].sort_values(ascending=False)
```

### Insight

- Size has the strongest impact on price
- Rooms also have a moderate effect on price
- Floor level and building height have weak influence on price

```python
plt.hist(df['price'], bins=50)
plt.title('Price Distribution')
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

## Size vs Price

This scatter plot shows the relationship between house size and price.

```python
plt.scatter(df['size'],df['price'],alpha=0.3)
plt.title('Size and Price')
plt.xlabel('size')
plt.ylabel('price')
plt.show()
```

### Insight

- There is a positive relationship between size and price
- Larger houses tend to have higher prices

## Rooms vs Price

This plot shows how the number of rooms affects house prices.

```python
plt.scatter(df['rooms'],df['price'],alpha=0.3)
plt.title('Rooms and Price')
plt.xlabel("Rooms")
plt.ylabel('Price')
plt.show()
```

### Insight

- There is a positive relationship between the number of rooms and price
- Houses with more rooms generally tend to have higher prices
- The relationship between the number of rooms and price is weaker compared to size

## Conclusion

- Price is strongly related to size
- Rooms have a moderate effect on price
- Location (district) also influences house prices
