# Tashkent House Price Prediction - Feature Engineering

## Objective
The goal of this notebook is to create new features, remove irrelevant ones,
and prepare the dataset for machine learning models.

```python
import pandas as pd
import numpy as np
```

```python
# Load cleaned dataset
df = pd.read_csv('../data/cleaned_data.csv')

df.head()
```

```python
# Drop unnecessary columns
df = df.drop(columns=['location'], errors='ignore')
```

```python
# Convert categorical variable into dummy variables
df = pd.get_dummies(df, columns=['district'], drop_first=True,dtype=int)
```

```python
df.head()
```

```python
# Define features and target
X = df.drop('price', axis=1)
Y = df['price']
```
