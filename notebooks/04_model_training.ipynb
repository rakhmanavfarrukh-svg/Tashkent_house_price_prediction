# Tashkent House Price Prediction - Model Training

## Objective
The goal of this notebook is to train and evaluate different machine learning models.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('../data/featured_data.csv')
df.head()
```

```python
X = df.drop('price',axis=1)
Y = df['price']
```

```python
# Split the dataset into training and testing sets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=77)
```

```python
# Scale features for models like SVM
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Linear Regression model
LR_model = LinearRegression()
LR_model.fit(X_train,Y_train)
# Make predictions
y_pred_LR = LR_model.predict(X_test)
```

```python
# Define a function to evaluate model performance
def evaluate(y_true, y_pred):
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R2:", r2_score(y_true, y_pred))
# Evaluate Linear Regression
evaluate(Y_test, y_pred_LR)
```

```python
# Train Random Forest model
RF_model = RandomForestRegressor(random_state=42)
RF_model.fit(X_train, Y_train)

# Make predictions
y_pred_RF = RF_model.predict(X_test)

# Evaluate model
evaluate(Y_test, y_pred_RF)
```

```python
# Train Support Vector Machine model 
svm = SVR()
svm.fit(X_train_scaled, Y_train)

# Make predictions
y_pred_svm = svm.predict(X_test_scaled)

# Evaluate model
evaluate(Y_test, y_pred_svm)
```

# Compare model performance based on evaluation metrics
# Typically, Random Forest performs best for this type of data

```python
# Save the best model
import joblib

joblib.dump(RF_model, '../models/random_forest_model.pkl')
```

```python
joblib.dump(scaler, '../models/scaler.pkl')
```
